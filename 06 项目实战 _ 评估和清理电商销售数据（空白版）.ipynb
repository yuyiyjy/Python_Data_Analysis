# 项目：评估和清理英国电商公司销售数据

## 分析目标

此数据分析的目的是，根据市场销售数据，挖掘畅销产品，以便制定更有效的市场策略来提升营收。

本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。

## 简介

原始数据集记录了一家英国在线零售公司在2010年12月1日至2011年12月9日期间的所有交易情况，涵盖了该公司在全球不同国家和地区的业务数据。该公司主要销售覆盖各个场景的礼品，包括但不限于生日礼品、结婚纪念品、圣诞礼品等等。该公司的客户群体主要包括批发商和个人消费者，其中批发商占据了相当大的比例。

数据每列的含义如下：
- `InvoiceNo`: 发票号码。6位数，作为交易的唯一标识符。如果这个代码以字母“c”开头，表示这笔交易被取消。
- `StockCode`: 产品代码。5位数，作为产品的唯一标识符。
- `Description`: 产品名称。
- `Quantity`: 产品在交易中的数量。
- `InvoiceDate`: 发票日期和时间。交易发生的日期和时间。
- `UnitPrice`: 单价。价格单位为英镑（£）。
- `CustomerID`: 客户编号。5位数，作为客户的唯一标识符。
- `Country`: 国家名称。客户所居住的国家的名称。

## 二.评估数据

### 随机取出多行，检查是否存在结构性问题

```python
original_data.sample(20)
```

### 经过取样发现并没有什么结构性问题，接下来按照步骤进行内容上的问题分析

### 1.评估缺失数据

```python
original_data.info()
```

### 通过info方法可以发现，Description和CustomerID两列数据非零行数小于总行数541909，所以这两列数据存在缺失值。另外可以发现InvoiceDate的数据类型为字符串而非日期类型，CustomerID为浮点数而非字符串类型

### 将Desription和CustomerID中存在缺失值的行取出来

```python
original_data[original_data["Description"].isnull()]
```

### 观察发现UnitPrice全为0，猜测是否Description缺失时UnitPrice为0；CustomerID也都为NaN

```python
original_data[(original_data["Description"].isnull()) & (original_data["UnitPrice"] != 0.0)]
```

### 发现Decription缺失的数据里UnitPrice全为0，这说明Description缺失时UnitPrice全为0

```python
original_data[(original_data["Description"].isnull()) & (original_data["CustomerID"].isnull())]
```

### 发现CustomerID也全为NaN，这说明Description缺失时UnitPrice全为0，CustomerID也全缺失，这部分数据需要删除，但要注意CustomerID缺失的数据要比Description缺失的数据更多，因此要删除的是Description缺失的那部分数据

### 结合“挖掘畅销产品”这一目标，由于产品名称全部缺失，因此这些数据对我们来说是无效数据，可以删除

### 2.评估重复数据

### 注意到数据里发票号码是交易的唯一标识符，产品代码是产品的唯一标识符，顾客ID是顾客的唯一标识符，但这些是允许重复的，因为同一件产品可能被购买多次，同一个顾客也可能多次购买产品

```python
original_data.duplicated(subset=["InvoiceNo","StockCode","CustomerID"])
```

### 3.评估不一致数据

### 对于数字类型的数据一般不会出现存在不同的表述的情况，而对于字符串有可能，但InvoiceNo、StockCode、Description这些数据正常也不会出现存在不同表述的情况，因此对于不一致数据我们只需要检查日期和国家

```python
original_data["InvoiceDate"].value_counts()
```

```python
original_data["Country"].value_counts().sort_index()
```

### 注意到日期没有不一致的情况，国家里英国有UK、U.K.、United Kingdom三种不同的表述，美国有USA、United States两种不同的表述

### 4.评估无效数据

```python
original_data.describe()
```

### 发现Quantity和UnitPrice的最小值均为负数，这是不合理的情况，而CustomerID此时依然为数字类型，一会儿清理数据时需要改成字符串类型

```python
original_data[original_data["Quantity"]<0]
```

### 观察发现展示出来的数据发票编号均为C开头，猜想是否Quantity为负数时发票编号均为C开头

```python
original_data[(original_data["Quantity"]<0) & (original_data["InvoiceNo"].str[0]!= "C")]
```

```python
original_data[(original_data["Quantity"]<0) & (original_data["InvoiceNo"].str[0]== "C")]
```

```python
original_data[(original_data["InvoiceNo"].str[0]== "C") & (original_data["Quantity"]>0)]
```

### 发现发票编号为C开头的数据Quantity全为负数

### 发现确实存在一些数据发票编号开头并不是C，又发现CustomerID均为NaN，所以接着检查CustomerID

```python
original_data[(original_data["UnitPrice"]<0)]
```

### 单价为负数只有两行数据，因此主要集中在Quantity为负数的情况上

```python
original_data[(original_data["Quantity"]<0) & (original_data["CustomerID"].isnull())]
```

### 发现Quantity为负数的数据中有一部分CustomerID缺失，但CustomerID对于数据分析并没有什么影响，所以主要考虑的是Quantity和InvoiceNo

```python
original_data[(original_data["Quantity"]<0) & (original_data["InvoiceNo"].str[0]!= "C") & (original_data["CustomerID"].isnull())]
```

## 三.清理数据

#### 1.将CustomerID数据类型转换为字符串
#### 2.将交易日期改为日期数据类型
#### 3.删除Description缺失的那部分数据
#### 4.将英国、美国的英文统一
#### 5.删除Quantity为负数的数据
#### 6.删除UnitPrice为负数的数据

### 首先创建一个新的DataFrame

```python
clean_data = original_data.copy()
```

```python
original_data.info()
```

### 1.将CustomerID数据类型转换为字符串

```python
clean_data["CustomerID"]=clean_data["CustomerID"].astype(str)
```

```python
clean_data.info()
```

```python
clean_data["CustomerID"]
```

### 但是转换后发现有小数点 我们不希望有小数点 所以需要进行切割

```python
clean_data["CustomerID"].str.slice(0,-2)
```

### 2.将交易日期改为日期数据类型

```python
clean_data["InvoiceDate"]=pd.to_datetime(clean_data["InvoiceDate"])
```

```python
clean_data["InvoiceDate"]
```

### 3.删除Description缺失的那部分数据

```python
clean_data=clean_data.dropna(subset=["Description"])
```

```python
print(clean_data["Description"].isnull().sum())
```

```python
clean_data
```

### 4.将英国、美国的英文统一

```python
clean_data=clean_data.replace({"U.K.":"United Kingdom","UK":"United Kingdom","USA":"United States"})
```

```python
clean_data["Country"].value_counts().sort_index()
```

### 5.删除Quantity为负数的数据

### 直接删除并不好删除，因为没学过筛选条件来删除的方法，但可以通过筛选出Quantity大于0的部分进行赋值来间接删除

```python
clean_data=clean_data[clean_data["Quantity"]>=0]
```

```python
clean_data[clean_data["Quantity"]<0]
```

### 6.删除UnitPrice为负数的数据

```python
clean_data=clean_data[clean_data["UnitPrice"]>=0]
```

```python
clean_data[clean_data["UnitPrice"]<0]
```

## 保存清理后的数据

```python
clean_data.to_csv("./e_commerce_cleand.csv",index=False)
```
